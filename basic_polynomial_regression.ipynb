```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
# X lies between -3 and +3, 200 cases
np.random.seed(102)
X = 6*(0.5-np.random.rand(200, 1))
```

```python
#y = 100*X**5 + 0.5*X**2 + 0.9*X + 4000*np.random.randn(200, 1)*0.5
y = 0.8*X**3 + 0.9*X + 2*np.random.randn(200, 1)
#visualize the data
plt.plot(X, y, 'b.')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("XY Graph")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)
```

```python
len(x_train), len(y_train), len(x_test), len(y_test)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print(r2_score(y_test, y_pred))
#len(y_pred)
```

```python
plt.plot(x_test, y_pred, "r.")
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#applying polynomial regression degree 3
poly = PolynomialFeatures(degree=3, include_bias=True)
x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.transform(x_test)
#include bias parameter
lr = LinearRegression()
lr.fit(x_train_trans, y_train)
y_pred = lr.predict(x_test_trans)
print(r2_score(y_test, y_pred))
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.plot(x_test, y_pred, "r.", label="Predictions")
#plt.plot(x_train, y_train, "b.",label='Training points')
plt.plot(X, y, "g.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
